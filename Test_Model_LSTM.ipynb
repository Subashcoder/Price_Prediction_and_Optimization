```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv(r"D:\Big Data Analytics- Sem 3\Capstone Project\Toronto\Milestone1&2\Columns_to_attach_calendar.csv")
df2=pd.read_csv(r"D:\Big Data Analytics- Sem 3\Capstone Project\Toronto\Milestone1&2\clean_calendar.csv")
```

```python
# List of columns to drop
columns_to_drop = ['available', 'adjusted_price', 'minimum_nights', 'maximum_nights']

# Drop the specified columns
df2 = df2.drop(columns=columns_to_drop)
```

```python
# Concatenate 'accommodates', 'property', and 'neighbourhood' columns into 'identifier_listing'
df1['identifier_listing'] = df1.apply(lambda row: f"{row['accommodates']}_{row['neighbourhood']}", axis=1)
```

```python
# Convert 'identifier_listing' column to 'category' dtype
df1['identifier_listing'] = df1['identifier_listing'].astype('category')

# Create a mapping dictionary
category_mapping = dict(enumerate(df1['identifier_listing'].cat.categories))

# Print the mapping dictionary
print(category_mapping)
```

```python
df1['identifier_listing'] = df1['identifier_listing'].astype('category').cat.codes
```

```python
df1
```

```python
# Merge df1 and df2 based on 'id' in df1 and 'listing_id' in df2
df = pd.merge(df1, df2, left_on='id', right_on='listing_id')

```

```python
df
```

```python
df=df.drop(["id", "accommodates", "neighbourhood"],axis=1)
```

```python
# Convert 'date' column to datetime dtype
df['date'] = pd.to_datetime(df['date'])

# Group by 'identifier_listing' and 'date' and calculate the mean of the other columns
df_1 = df.groupby(['identifier_listing', 'date']).mean().reset_index()

```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import numpy as np

# Assuming df is your DataFrame and 'price' is the column with the prices
train_size = int(len(df_1) * 0.8)
train, test = df_1[0:train_size], df_1[train_size:]

model = ARIMA(train['price'], order=(0,1,1))
model_fit = model.fit()

forecast = model_fit.predict(start=len(train), end=len(train)+len(test)-1)

# Evaluate the performance of the model on the test set
mse = mean_squared_error(test['price'], forecast)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test['price'], forecast)
mape = mean_absolute_percentage_error(test['price'], forecast)

print('MSE:', mse)
print('RMSE:', rmse)
print('MAE:', mae)
print('MAPE:', mape)
```
