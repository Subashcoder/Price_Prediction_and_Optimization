```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Big Data Analytics- Sem 3\Capstone Project\Toronto\calendar.csv", index_col=False)
```

```python
def price_cleaning(df,column):
    df[column] = pd.to_numeric(df[column].apply(lambda x: str(x).replace('$', '').replace(',', '')), errors='coerce')
    return df[column].head()
```

```python
# Converting the price in calendar_df
price_cleaning(df,'price')
```

```python
df['adjusted_price'] = df['adjusted_price'].fillna(0)
```

```python
# Converting the price in calendar_df
price_cleaning(df,'adjusted_price')
```

```python
# replace 't' with 1 and else with 0 in the 'available' column
df['available'] = df['available'].replace({'t': 1, 'f': 0})
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
df.info()
```

```python
df_1=df
```

```python
# Group by 'listing_id' and compute min and max dates
date_ranges = df_1.groupby('listing_id')['date'].agg(['min', 'max']).reset_index()

# Rename the columns for clarity
date_ranges.rename(columns={'min': 'start_date', 'max': 'end_date'}, inplace=True)
```

```python
df_1.info()
```

```python
# Step 6: Merge date_ranges into df
print("Merging date_ranges into the original DataFrame...")
df_enriched = pd.merge(df_1, date_ranges, on='listing_id', how='left')
print(f"Enriched DataFrame shape: {df_enriched.shape}")
```

```python
df_enriched
```

```python
# delete the date column
df_enriched = df_enriched.drop(['date', 'available'], axis=1)

# delete duplicate rows
df_enriched = df_enriched.drop_duplicates()

# reset the index
df_enriched = df_enriched.reset_index(drop=True)
```

```python
df_enriched
```

```python
# Find rows where listing_id is being repeated
duplicate_rows = df_enriched[df_enriched.duplicated(subset='listing_id', keep=False)]

# Display the duplicate rows
print(duplicate_rows)
```

```python
# List of columns to drop
columns_to_drop = ['adjusted_price', 'minimum_nights', 'maximum_nights']

# Drop the specified columns
df_enriched = df_enriched.drop(columns=columns_to_drop)

# delete duplicate rows
df_enriched = df_enriched.drop_duplicates()

# reset the index
df_enriched = df_enriched.reset_index(drop=True)
```
